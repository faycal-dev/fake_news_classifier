for the text we can do 2 operations :
    (stamming : delet all the suffixes and prefixes from text expl: studies -> studi) 
    (limmatizatioon : do the same thing but returns more significant words : studies -> study)

```python
from nltk.stem.porter import PorterStemmer
porter_stemmer  = PorterStemmer()
text = "studies studying cries cry"
tokenization = nltk.word_tokenize(text)
for w in tokenization:
    print("Stemming for {} is {}".format(w,porter_stemmer.stem(w))) 
```

```python
from nltk.stem import WordNetLemmatizer
wordnet_lemmatizer = WordNetLemmatizer()
text = "studies studying cries cry"
tokenization = nltk.word_tokenize(text)
for w in tokenization:
    print("Lemma for {} is {}".format(w, wordnet_lemmatizer.lemmatize(w)))
```

```python
useless_words = stopwords.words('english')
useless_words
```

```python
from nltk.stem.porter import PorterStemmer
from nltk.stem import WordNetLemmatizer

ps = PorterStemmer()
wl = WordNetLemmatizer()

corpus_stem = []
corpus_limm = []
for i in range(0, len(message)):
#     we leave only the alphanumeric letters
    review = re.sub('[^a-zA-Z0-9]', ' ', message['title'][i])
    review = review.lower()
    review = review.split()
### here we can use any one ofe the two methodes
#     stem every word that is not in the usless words
    review_stem = [ps.stem(word) for word in review if not word in useless_words]
    review_stem = ' '.join(review_stem)
    corpus_stem.append(review_stem)
#     limm every word that is not in the usless words
    review_limm = [wl.lemmatize(word) for word in review if not word in useless_words]
    review_limm = ' '.join(review_limm)
    corpus_limm.append(review_limm)
```

```python
corpus_stem
```

```python
corpus_limm
```

```python
from tensorflow.keras.preprocessing.text import Tokenizer
```

```python
# transforming the texts to a vector
tokenizer = Tokenizer(oov_token='<unknown>')
tokenizer.fit_on_texts(corpus_limm)
```

```python
# every single word is in our word index
word_index = tokenizer.word_index
print(word_index)
```

```python
print(f"the lenght of our vocabulary is {len(word_index)} instead of 31803 if we did not use limmetizing and deleting usless words")
```

```python
vectorized_text = np.array(tokenizer.texts_to_sequences(corpus_limm))
vectorized_text[:5]
```

```python
# or we can use one hot encoding
from tensorflow.keras.preprocessing.text import one_hot

onehot_repr=[one_hot(words,len(word_index))for words in corpus_limm] 
onehot_repr = np.array(onehot_repr)
onehot_repr[:5]
```

```python
max_lenghth = 0
for i in range(len(vectorized_text)):
    if (len(vectorized_text[i]) > max_lenghth):
        max_lenghth = len(vectorized_text[i])
print(f"the maximum length of a sentence is {max_lenghth}")
```

```python
max_lenghth = 0
for i in range(len(onehot_repr)):
    if (len(onehot_repr[i]) > max_lenghth):
        max_lenghth = len(vectorized_text[i])
print(f"the maximum length of a sentence is {max_lenghth}")
```

```python
# so we need to add padding to all the sentences so that they will have the same size to give for the model
from tensorflow.keras.preprocessing.sequence import pad_sequences
pad_sentences = pad_sequences(vectorized_text, padding='post',maxlen=max_lenghth)
# or we do it with the onehot encoding
onehot_pad_sentence = pad_sequences(onehot_repr, padding='post',maxlen=max_lenghth)
pad_sentences[:5]
```

the model

```python
from tensorflow.keras.layers import Embedding
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import Bidirectional
from tensorflow.keras.layers import Dropout
```

```python
# the dropout is added to prevent overfitting
vocab_size = len(word_index)
embeded_vector_size = 50
model = Sequential()
model.add(Embedding(vocab_size,embeded_vector_size, input_length=max_lenghth))
model.add(Dropout(0.3))
model.add(LSTM(128))
model.add(Dropout(0.3))
model.add(Dense(1,activation='sigmoid'))
```

```python
model.compile(loss='binary_crossentropy', optimizer="adam", metrics=['accuracy'])
```

```python
print(model.summary())
```

```python
from sklearn.model_selection import train_test_split
y = np.array(y)
X_train, X_test, y_train, y_test = train_test_split(onehot_pad_sentence, y, test_size=0.3)
```

```python
model.fit(X_train, y_train, epochs=8, batch_size=64)
```

```python
model.evaluate(X_test, y_test)
```
